```python
class AngleMeasurer:
    def __init__(self, filepath:str):
        self.img = cv2.imread(filepath)
        self.pointList = []
        self.filepath = filepath

    def mouse_click(self, event, x, y, flags, params):
        if event == cv2.EVENT_LBUTTONDOWN:
            cv2.circle(self.img, (x,y), 5, (0, 0, 255),cv2.FILLED)
            size = len(self.pointList)
            color = (np.random.randint(0 ,255), np.random.randint(0 ,255), np.random.randint(0 ,255))
            if size!=0 and size % 3 != 0:
                cv2.line(self.img, tuple(self.pointList[round((size - 1)/3)*3]), (x, y),color, 3)
            self.pointList.append([x, y])
    def gradient(self,pt1, pt2):
        return ((pt2[1] - pt1[1])/ (pt2[0] - pt1[0]))
    def getangle(self):
        """
        tan θ = |(m2 - m1)/ (1 + m1.m2)|
        """
        pt1, pt2, pt3 = self.pointList[-3:]
        m1 = self.gradient(pt1, pt2)
        m2 = self.gradient(pt1, pt3)
        
        thetaR = math.atan((m2 - m1)/ (1 + (m1*m2)))
        thetaD = int(abs(round(math.degrees(thetaR))))
    
        cv2.putText(self.img, str(thetaD), (pt1[0]+10, pt1[1]+10), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255),2)
        return thetaD
    def process(self):
        while True:
            cv2.imshow("Image", self.img)
            cv2.setMouseCallback("Image", self.mouse_click)
            if len(self.pointList)%3 == 0 and len(self.pointList)>0:
                angle = self.getangle()
            if (cv2.waitKey(1) & 0xFF == ord("q")):
                self.pointList = []
                self.img = cv2.imread(self.filepath)
            if cv2.waitKey(1) & 0xFF == ord("e"):
                break
    cv2.destroyAllWindows()

        
```

```python
measure_angle = AngleMeasurer("laptop.png")
measure_angle.process()
```
